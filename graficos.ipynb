```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Gráficos para os resultados do modelo Sarsa com randomização das envs

```python
dados_sarsa = pd.read_csv("states/2023-11-12T17:40:44/log.csv", names=["indice", "reward", "loss"])
```

```python
# lista dos modelos números a cada 500 episódios
lista_models = []
for i in range(0, len(dados_sarsa), 500):
    lista_models.append(i)

```

```python
# Valores de recompensa com indice da lista_models
rewards = []
for i in lista_models:
    rewards.append(dados_sarsa["reward"][i])

# Maior valor de reward
max_reward = max(rewards)
print("Maior valor de reward dos modelos salvos: ", max_reward)

# Indice do max_reward
indice_max_reward = rewards.index(max_reward)
print("Indice do max_reward: ", indice_max_reward)

indice_max_reward = lista_models[indice_max_reward]
print("Indice do max_reward nos modelos: ", indice_max_reward)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(dados_sarsa["reward"])
plt.xlabel("Episódio")
plt.ylabel("Reward")
plt.title("Reward por episódio")

plt.subplot(1, 2, 2)
plt.plot(dados_sarsa["loss"])
plt.xlabel("Episódio")
plt.ylabel("Loss")
plt.title("Loss por episódio")

plt.tight_layout()
plt.show()
```
